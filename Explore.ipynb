```python
plt.figure(figsize=(12,8))
df1 = df[df['minimum_nights']==1]
df2 = df1.groupby(['room_type','neighbourhood_group'])['price'].mean().sort_values(ascending=True)
df2.plot(kind='bar')
plt.title('Average Price for rooms in neighbourhood group')
plt.ylabel('Average Daily Price')
plt.xlabel('Neighbourhood Group')
plt.show()
```

```python
def category_price(price):
  if price <= 69:
    return '25%'
  elif price <= 106:
    return '50%'
  elif price <= 175:
    return '70%'
  else:
    return '100%'
```

```python
df['category_price'] = df['price'].apply(category_price)
```
